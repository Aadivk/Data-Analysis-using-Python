```python
#convert

data['Date'] = pd.to_datetime(data['Date'])
```

```python
## Highest temperature registered
max_temp=data.groupby('Country')['Average Temperature'].mean().round(2)
max_temp_sort=max_temp.sort_values(ascending=False).reset_index()
max_temp_sort.head(10)
```

```python
# BarPlot for Highest temperature by country
plt.figure(figsize = (10,15))
k=sns.barplot(x=max_temp_sort['Average Temperature'],y=max_temp_sort['Country'],palette = 'muted')
k.bar_label(k.containers[0], fontsize = 8)
plt.title('Highest temperature by country')
plt.show()
```

```python
## Lets check the Lowest temperature data registered
min_temp=data.groupby('Country')['Average Temperature'].mean().round(2)
min_temp_sort=min_temp.sort_values(ascending=True).reset_index()
k=min_temp_sort.head(10)
k
```

```python
# Lets check top 10 countries with Lowest temperature data registered using seaborn barplot
plt.figure(figsize=(13,7))
lowest=plt.bar(k['Country'],k['Average Temperature'],color='lightgreen',width=0.5)
plt.xlabel('Country')
plt.ylabel('Average Temperature')
plt.title("Top 10 countries with Lowest temperature")
plt.grid(True)
plt.show()
```

```python
## Hottest Cities
hottest_city=data.groupby('City')['Average Temperature'].mean().round(2)
hottest_city_sort=hottest_city.sort_values(ascending = False).reset_index()
day=hottest_city_sort.head(21)
day
```

```python
# showing 20 hottest cities using barplot 
plt.figure(figsize=(16,8))
city=sns.barplot(x=day['Average Temperature'],y=day['City'],palette='muted')
city.bar_label(city.containers[0],fontsize=10)

plt.title('Average Temperature by City')
plt.show()
```

```python
# coolist city 
cool_day=data.groupby('City')['Average Temperature'].mean().round(2)
cool_day_sort=cool_day.sort_values(ascending = True).reset_index()
cool=cool_day_sort.head(11)
cool
```

```python
# Showing top 10 coldest cities
plt.figure(figsize=(10,8))
plt.plot(cool['Average Temperature'],cool['City'],color='green',marker='D',mfc='k',ms=8,mec='r')
plt.xlabel("Average Temperature")
plt.ylabel('Cities')
plt.title("Showing top 10 lowest Average Temperature cities")
plt.grid(True)

plt.show()
```

```python
# Lets check the Temperature over time
max_time=data.groupby('Date')['Average Temperature'].mean().round()
max_sort=max_time.sort_values(ascending=False).reset_index()
max_sort.head()
```

```python
# check temperatur by time using lineplot seaborn 
plt.figure(figsize=(16,8))
sns.lineplot(data=max_sort,x=max_sort['Date'],y=max_sort['Average Temperature'])
plt.title("Average Temperature Over Time")
plt.show()
```
