```python
# prompt: visualization chart using mat and sea

import matplotlib.pyplot as plt
import seaborn as sns

# Visualize the distribution of the 'quality' variable
plt.figure(figsize=(8, 6))
sns.countplot(x='quality', data=df)
plt.title('Distribution of Wine Quality')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.show()

# Create a correlation matrix heatmap
plt.figure(figsize=(12, 10))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Wine Quality Features')
plt.show()

# Visualize the relationship between 'alcohol' and 'quality'
plt.figure(figsize=(8, 6))
sns.scatterplot(x='alcohol', y='quality', data=df)
plt.title('Alcohol Content vs. Wine Quality')
plt.xlabel('Alcohol Content')
plt.ylabel('Quality')
plt.show()

# Visualize the relationship between 'residual sugar' and 'quality'
plt.figure(figsize=(8,6))
sns.boxplot(x='quality', y='residual sugar', data=df)
plt.title('Residual Sugar vs Wine Quality')
plt.xlabel('Quality')
plt.ylabel('Residual Sugar')
plt.show()

# Visualize the distribution of 'pH' for different quality levels
plt.figure(figsize=(8, 6))
sns.kdeplot(df.loc[df['quality'] == 3, 'pH'], label='Quality 3', shade=True)
sns.kdeplot(df.loc[df['quality'] == 4, 'pH'], label='Quality 4', shade=True)
sns.kdeplot(df.loc[df['quality'] == 5, 'pH'], label='Quality 5', shade=True)
sns.kdeplot(df.loc[df['quality'] == 6, 'pH'], label='Quality 6', shade=True)
sns.kdeplot(df.loc[df['quality'] == 7, 'pH'], label='Quality 7', shade=True)
sns.kdeplot(df.loc[df['quality'] == 8, 'pH'], label='Quality 8', shade=True)
plt.xlabel("pH")
plt.ylabel("Density")
plt.title("Distribution of pH by Wine Quality")
plt.legend()
plt.show()
```

**Define y & x**

```python
y = df['quality']
```

```python
y.shape
```

```python
y
```

```python
x = df[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
        'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
        'pH', 'sulphates', 'alcohol']]
```

```python
x = df.drop(['quality'], axis=1)
```

```python
x.shape
```

```python
x
```

**Get X Variables Standardized**

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
x = ss.fit_transform(x)
```

```python
x
```

**Get Train Test Split**

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test, = train_test_split(x,y, test_size = 0.3, stratify = y, random_state=222529)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

**Get Model Train**

```python
from sklearn.svm import SVC
```

```python
svc = SVC()
```

```python
svc.fit(x_train, y_train)
```

**Get Model Prediction**

```python
y_pred = svc.predict(x_test)
```

```python
y_pred.shape
```

```python
y_pred
```

**Get Model Evaluation**

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```
